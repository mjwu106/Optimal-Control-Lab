# <font color=blue> Question 1.  Equilibrium Point and Linearization </font>

<center></center>

Consider an inverted double-pendulum system that consists of two point masses $m=2$ kg on two rods of length $l=6$ m. The rods are supported in a pivot on a horizontal plane with $g = 9.81$ m/s$^2$. At length $a=3$ m they are connected by a spring-damper system. The spring has a spring constant $k = 1.5$ N/s. The damping coefficient is $d = 1$ Ns/m. Its neutral position is at $\theta_1(t)=\theta_2(t)=0$. In the pivot of the left mass, an external torque $T(t)$ can be applied. In the right pivot there is a disturbance torque of the form $\alpha(t)+\beta\dot{\theta}_2^2$, where $\beta = 0.1$ m$^{2}$kg is a constant.

The differential equations representing the dynamics of the system are

\begin{align}
ml^2\ddot{\theta}_1(t) &= mgl\sin\theta_1(t) +a\cos\theta_1(t) F_{sd}+T(t)\nonumber\\
ml^2\ddot{\theta}_2 (t) &= mgl\sin\theta_2(t) -a\cos\theta_2(t)  F_{sd}+\alpha(t)+\beta\dot{\theta}_2^2(t) \nonumber
\end{align}

where $F_{sd}=k\:a(\sin\theta_2(t)  -\sin\theta_1(t) )+d\:a(\dot{\theta}_2(t) -\dot{\theta}_1(t) )$ which gives the following set of differential equations

\begin{align}
\ddot{\theta}_1 (t) &= \frac{1}{ml^2}\left[mgl\sin\theta_1 (t)+a^2\cos\theta_1(t)\left(k(\sin\theta_2(t) -\sin\theta_1(t))+d(\dot{\theta}_2(t)-\dot{\theta}_1(t))\right)+T(t)\right]\nonumber\\
\ddot{\theta}_2 (t)&= \frac{1}{ml^2}\left[mgl\sin\theta_2(t)-a^2\cos\theta_2(t)\left(k(\sin\theta_2(t) -\sin\theta_1(t))+d(\dot{\theta}_2(t)-\dot{\theta}_1(t))\right)+\alpha(t)+\beta\dot{\theta}_2^2(t)\right]\nonumber
\end{align}

**Part (a)**

For any fixed values of $\theta_2$  and $\alpha$ (denoted ${\bar \theta}_2$ and $\bar \alpha$), there
exist unique values of 
$T$ and $\theta_1$ ($\bar T$ and ${\bar \theta}_1$) such the system is in equilibrium (with ${\dot \theta}_1 = {\dot \theta}_2 = 0$, of course).  Write a function $\texttt{EquilPoint}$, with function declaration:

`def EquilPoint(theta2_bar, alpha_bar):
    return  theta1_bar, T_bar `

that carries out this calculation.
All the input and output arguments ($\texttt{alpha_bar}$, $\texttt{theta2_bar}$, $\texttt{theta1_bar}$ and $\texttt{T_bar}$) are scalar valued.

### <font color=red> Deliverable 1a: write your answer in the code cell below</font>


```python
import numpy as np
from numpy import sin, cos, tan, arcsin 
m=2
l=6
g=9.81
a=3
k=1.5
d=1
beta=0.1

def EquilPoint(theta2, alpha):
    # Write your code here:
    theta1 = arcsin((-alpha-m*g*l*sin(theta2))/(a**2*cos(theta2)*k)+sin(theta2))
    T = -m*g*l*sin(theta1)-a**2*cos(theta1)*(k*sin(theta2)-k*sin(theta1))
    return  theta1, T


# Please submit the results of the following unit test
theta1_bar, T_bar = EquilPoint(theta2=0.1, alpha=0.5)
print(theta1_bar)
print(T_bar)
```

**Part (b)** 

Next we want to compute the Jacobian linearization about such an equilibrium point. Denote
\begin{align}
\bar x := \left[ \begin{array}{c}
{\bar \theta}_1 \\ {\bar {\dot \theta}}_1 \\
{\bar \theta}_2 \\ {\bar {\dot \theta}}_2 \end{array} \right], \ \ \ \ \ 
\bar u := \left[ \begin{array}{c}
{\bar T} \\ {\bar \alpha} \end{array}  \right], \ \ \ \
\bar y = {\bar \theta}_1
\end{align}

Deviations from equilibrium point are denoted $x$, $u$ and $y$, namely

\begin{align}
x(t) := \left[ \begin{array}{c} \theta_1 (t) \\ {\dot \theta}_1 (t) \\ \theta_2 (t) \\ {\dot \theta}_2 (t) \end{array}
\right] - \bar x, \ \ \ \ \
u(t) := \left[ \begin{array}{c} T (t) \\ \alpha (t)  \end{array}
\right] - \bar u, \ \ \ \ \
y(t) := \theta_1 (t) - \bar y
\end{align}

and, to first-order, are governed by linear differential equations of the form
	    \begin{align}
	    	\dot{x}(t) &= Ax(t)+Bu(t)\nonumber\\
	    	y(t)       &= Cx(t)+Du(t)\nonumber
	    \end{align}
where the matrices $A, B, C$ and $D$ depend on the equilibrium point and the underlying nonlinear differential equations
for the system.

Write a function called $\texttt{LinearizeModel}$ with function declaration line:

`def LinearizeModel(theta2_bar, alpha_bar):
    return A, B, C, D, theta1_bar, T_bar`

that carries out this calculation at any given value of $({\bar \theta}_2, \bar \alpha)$.  This function, $\texttt{LinearizeModel}$ should call the equilibrium point function $\texttt{EquilPoint}$
to determine ${\bar \theta}_1$ and $\bar T$ and also return these as the 5th and 6th output arguments.

### <font color=red> Deliverable 1b: write your answer in the code cell below</font>

```python
import sympy as sym 
import numpy as np
# Hint: use SymPy to compute jacobians.

def LinearizeModel(theta2, alpha):
    # Write your code here:\
    theta1, T = EquilPoint(theta2, alpha)
    x1, x2, x3, x4, u1, u2 = sym.symbols('x1 x2 x3 x4 u1 u2') 
    Asub = [x2,(m*g*l*sym.sin(x1)+a**2*sym.cos(x1)*(k*(sym.sin(x3)-sym.sin(x1))+d*(x4-x2))+u1)*(1/(m*l**2)),x4,(m*g*l*sym.sin(x3)-a**2*sym.cos(x3)*(k*(sym.sin(x3)-sym.sin(x1))+d*(x4-x2))+u2+beta*x4**2)/(m*l**2)]
    Asub=sym.Matrix(Asub).jacobian([x1, x2, x3, x4]).subs([(x1,theta1),(x2,0),(x3,theta2),(x4,0)])
    Bsub = [x2,(m*g*l*sym.sin(x1)+a**2*sym.cos(x1)*(k*(sym.sin(x3)-sym.sin(x1))+d*(x4-x2))+u1)/(m*l**2), x4, (m*g*l*sym.sin(x3)-a**2*sym.cos(x3)*(k*(sym.sin(x3)-sym.sin(x1))+d*(x4-x2))+u2+beta*x4**2)/(m*l**2)]
    Bsub = sym.Matrix(Bsub).jacobian([u1, u2]).subs([(u1,T),(u2,alpha)])
    C = sym.Matrix([x1]).jacobian([x1, x2, x3, x4])
    D = sym.Matrix([x1]).jacobian([u1, u2])
    return Asub, Bsub, C, D, theta1, T


# Please submit the results of the following unit test
theta2_bar = 0.1
alpha_bar = 0.5
A, B, C, D, theta1_bar, T_bar = LinearizeModel(theta2_bar, alpha_bar)
print('A = ', A)
print('B = ', B)
print('C = ', C)
print('D = ', D)
print('theta1_bar = ', theta1_bar)
print('T_bar = ', T_bar)
```

# <font color=blue> Question 2. Euler Discretization of a Building Heat Transfer Model
Consider the nonlinear heat transfer dynamics building room temperature with forced air ventilation:
\begin{align}
m_z c_z \dot{T}(t) = q(t) + c_pu_1(t)(u_2(t) - T(t)) 
\end{align}
where

\begin{equation}
\begin{aligned}
T(t) &= \text{room temperature} \nonumber \\
u_1(t) &= \text{air mass flow rate} \nonumber \\
u_2(t) &= \text{supply air temperature} \nonumber.\\
q(t) &= \text{ heat load} \nonumber \\
m_z &= \text{effective room mass (constant)}  \nonumber \\
c_z &= \text{effective room heat capacity (constant)} \nonumber \\
c_p &= \text{heat capacity of air (constant)} \nonumber \\
%u &= \text{ heat power input}
\end{aligned}
\end{equation}

**Part (a)** 

Write a function $\texttt{bldgHTM}$, with function declaration line:

`def bldgHTM(T, u1, u2, q, mz, cz, cp):
    return Tdot`

that implements this model, returning $\dot T(t)$, given the values $T(t), u_1 (t), u_2 (t), q(t)$ and the values for the
3 fixed parameters.

Note that for this system, the state of the system is the temperature $T$ of the room, the input $u_1$ is airflow rate (usually controlled through a fan), the input $u_2$ is the temperature of the air entering the room (usually controlled with a heating or cooling coil). Everything else is a model parameter.
For this exercise we will use the values $T_s=0.1$, $c_p= 1000$, $m_z= 100$ and $c_z=20$.



### <font color=red> Deliverable 2a: write your answer in the code cell below </font>

```python
def bldgHTM(T, u1, u2, q, mz, cz, cp):
    # Write your code here:
    Tdot = (q+cp*u1*(u2-T))/(mz*cz)
    return Tdot

# Please submit the results of the following unit test
Tdot = bldgHTM(T=60, u1=0.001, u2 = 10, q=1000,  mz=100, cz=20, cp=1000)
print(Tdot)
```

**Part (b)** 

Do a Forward Euler Discretization to take this model from continuous time to discrete time. Assume you are given the discrete time load $q(k)$.
Write a function:

`def eulerDiscretization(T,q,u1,u2):
    return T_KplusOne`

to get $T(k+1)$ as a function of the inputs $u_1(k), u_2(k)$, state $T(k)$, load $q(k)$. $\textbf{Note:}$ All the input and output arguments ($\texttt{T}$, $\texttt{q}$, $\texttt{u1}$, $\texttt{u2}$ and $\texttt{T}_\texttt{KplusOne}$) are scalars.

### <font color=red> Deliverable 2b: write your answer in the code cell below</font>

```python
def eulerDiscretization(T,q,u1,u2):
    mz = 100
    cz = 20
    Ts = 0.1
    cp = 1000
    # Write your code here:
    T_KplusOne = T*(1-cp*u1*Ts/(mz*cz))+Ts*(q+cp*u1*u2)/(mz*cz)
    return T_KplusOne

# Please submit the results of the following unit test
T_KplusOne = eulerDiscretization(T=60, q=1000, u1=0.001, u2 = 100)
print(T_KplusOne)
```

\dot{x} &= v \cos(\psi+\beta)\\

l_f &= \text{ distance from the center of mass of the vehicle to the front axle} \nonumber\\
\delta_f &= \text{ steering angle of the front wheels with respect to the longitudinal axis of the car} \nonumber

```python
import matplotlib.pyplot as plt

def carModel(beta, a, x, y, v, psi): 
    l_r   = 1.738
    dt = 0.1

    # Write your code here: 
    x_out = x+dt*v*cos(psi+beta)
    y_out = y+dt*v*sin(psi+beta)
    v_out = v+a*dt
    psi_out = psi+v/l_r*sin(beta)*dt
    return x_out, y_out, v_out, psi_out


# Please submit the results of the following unit test
x, y, v, psi = carModel(0.1,2,5,2,0.1,10)
print(x)
print(y)
print(v)
print(psi)
```
